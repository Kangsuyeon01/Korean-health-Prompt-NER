```python
# Mecab 객체 생성
mecab = Mecab()

# 형태소 태깅 함수
def morpheme_tagging(text):
    node = mecab.pos(text)
    node = ["("+ str(i[0]) +','+ str(i[1])+")" for i in node]
    return " / ".join(node)
```

```python
# 형태소 태깅 수행 및 'pos' 열에 추가
df['pos'] = df['sentences'].apply(morpheme_tagging)
```

```python
text = """ Defn: 개체명 태그는 각각 BodyPart, Symptom, Disease가 있다. 
각각 BIO 태깅 기법으로 각 개체에 대해 BodyPart는 B-BD 와 I-BD, Symptom는 B-SP와 I-SP, Disease는 B-DS와 I-DS로 정의한다.
문장은 공백 단위로 분리한다. BIO가 O인 경우에는 Entity False로 처리하고, 그 외는 True로 처리한다.

Q: 주어지는 문장에 대해 샘플을 참고하여 Input Sentence에대해 BIO 태그를 이용하여 Sample과 동일하게 output을 구성하여라

Sample:
- sample sentence: 어지럼증이 걱정되어 질문하신 것 같습니다.
- sample POS: 어지럼증(NNG) / 이(JKS) / 걱정(NNG) / 되(XSV) / 어(EC) / 질문(NNG) / 하(XSV) / 신(EP+ETM) / 것(NNB) / 같(VA) / 습니다(EF) / .(SF) / 
- sample BIO: B-ST O O O O O
- sample output:
Word | POS | Entity | BIO 
어지럼증이 |  어지럼증(NNG) + 이(JKS) | True | as it is Symptom (B-ST)
걱정되어 | 걱정(NNG) + 되(XSV) + 어(EC) | False | as it is a verb (O)
질문하신 | 질문(NNG) + 하(XSV) + 신(EP+ETM) | False | as it is a verb (O)
것 | 것(NNB) | False | as it is not a named entity (O)
같습니다 | 같(VA) / 습니다(EF) | False| as it is a verb(O)
. | .(SF) | False | as it is a preiod (O) 

Input:
- input sentence: 어지럼증이 1초간 나타났다면 아찔한 느낌 정도가 아니었나 생각됩니다.
- input POS: 어지럼증(NNG) / 이(JKS) / 1(SN) / 초간(NNG) / 나타났(VV+EP) / 다면(EC) / 아찔(XR) / 한(XSA+ETM) / 느낌(NNG) / 정도(NNG) / 가(JKC) / 아니(VCN) / 었(EP) / 나(EC) / 생각(NNG) / 됩니다(XSV+EF) / .(SF) / 
- input BIO: B-ST O O O O O O O O
- output: """
```

```python
def word_pos_match(sentence):
    t = sentence.split(" ")
    t_s = mecab.pos(sentence)
    t_dict = {i:[] for i in range(len(t))}

    cnt = 0
    for i in range(len(t)):
        tmp_text = ""
        while cnt < len(t_s):
            if tmp_text != t[i]:
                if t_s[cnt][0] in t[i]:
                    tmp_text += t_s[cnt][0]
                    t_dict[i].append(t_s[cnt])
                    cnt += 1
                else:
                
                    break
            else:
                break
    return t_dict
```

```python
def format_structure(sentence, pos, bio):
    text = []
    s = sentence.split(" ")
    match_dict = word_pos_match(sentence) # 공백단위 단어 분절과 포스태깅 결과 매칭 데이터
    b = bio.split(" ")
    bio_s = []

    for i in b:
        if i == "O":
            bio_s.append("False")
        else:
            bio_s.append("True")
         
    text.append('Word | POS | Entity | BIO ')
    
    tag_dict = {"B-ST":"Symptom","I-ST":"Symptom","B-BD":"BodyPart","I-BD":"BobyPart","B-DS":"Disease","I-DS":"Disease"}
    pos_dict = {'NNG':'개체명 범주에 속하지 않은 명사이기 때문임','NNP':'개체명 범주에 속하지 않은 명사이기 때문임','NNB':'개체명 범주에 속하지 않은 명사이기 때문임',
                'VV':'용언이기 때문임','VA':'용언이기 때문임','VS':'보조용언이기 때문임','MM':'관형사이기 때문임',
                'JKS':'조사이기 때문임','JKC':'조사이기 때문임','JKG':'조사이기 때문임','JKO':'조사이기 때문임','JKB':'조사이기 때문임','JKV':'조사이기 때문임','JKQ':'조사이기 때문임','JX':'조사이기 때문임','JC':'조사이기 때문임',
                'SF':'부호이기 때문임','SE':'부호이기 때문임','SSO':'부호이기 때문임','SSC':'부호이기 때문임','SC':'부호이기 때문임','SY':'부호이기 때문임','SW':'부호이기 때문임'}
    for i in range(len(s)):
        match_pos = "+".join([f"{token[0]}({token[1]})" for token in match_dict[i]])
        check_pos = [token[1] for token in match_dict[i]]
        if bio_s[i] == "True":
            text.append(f"{s[i]} |{match_pos}| {bio_s[i]} | {b[i]} ")
        else:
            
            # OOV 의 이유 적기
            # 형태소 태그를 근거로 원인 적기
            # verve , noun , ... 등에 대한 이유 적기
            reason = ""
            for pp in check_pos:
                if pp in pos_dict:
                    reason += pos_dict[pp]
                    break
            if reason == "":
                reason = "개체명 범주에 속하지 않는 단어임"
            text.append(f"{s[i]} |{match_pos}| {bio_s[i]} | {b[i]} ")
        
    return "\n".join(text)

df['format'] = df.apply(lambda row: format_structure(row['sentences'], row['pos'], row['bio_tag']), axis=1)
```
